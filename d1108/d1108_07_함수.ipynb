```python
import pandas as pd

df=pd.read_csv('score.csv',index_col='지원번호')
```

함수 선언 호출: apply()

```python
def add_cm(height):
  return str(height) + 'cm'

```

```python
# 함수 호출:apply()
df['키']=df['키'].apply(add_cm)
df
```

```python
df['SW특기'] = df['SW특기'].str.lower()
```

```python
# 함수 선언
def change(lang):
  if pd.notnull(lang):
    return lang.upper()
  else :
    return 'PYTHON'
```

```python
df['SW특기'].apply(change)
```
