```python
import pandas as pd

df= pd.read_excel('score.xlsx',index_col='지원번호')
```

데이터 수정

```python
## replace : 데이터 값을 변경, inplace =True 지정
df['학교'].replace({'구로고':'가산고'},inplace=True)
df['학교'].replace({'구로고':'가산고','디지털고':'단지고'})

# 특정 컬럼 값 변경
# 컬럼 = 변경 값 입력
df['SW특기'] = df['SW특기'].str.lower() # 소문자
df['SW특기'].str.upper() # 대문자
df['SW특기'].str.capitalize() # 첫글자만 대문자
df['키']=200

# 특정 셀값 변경
df.loc['1번','국어']=100
df.loc['2번',['국어','영어']] =10
df.loc[['2번','3번'],['국어','영어']] =50

# 컬럼명=문자+문자
df['학교']=df['학교']+'등학교'
```

```python
df['학교']=df['학교']+'등학교'
df
```

```python
df.loc[['2번','3번'],['국어','영어']] =50
```

데이터 추가
- 컬럼 추가 : df[컬럼명]
- row 추가 : df.loc[index명]

```python
## 데이터 추가
## 컬럼에 없는 컬럼 이름을 입력하면 컬럼이 추가 됨.
df['합계'] = 0
df['합계'] = df['국어']+df['영어']+df['수학']+df['사회']
df['평균'] = df['합계']/4
df
```

조건식을 사용해서 값 변경


```python
df['결과']='Fail'
# 조건식을 사용해서 값을 변경

# 결과 컬럼에서 조건식을 사용해서 값 변경
filt = (df['합계']>=300)
df.loc[filt,'결과'] = 'pass'
```

```python
df['합산']=100
# 국어 점수가90점 이상이면 합산에 +100점을 더해라
filt = df['국어']>=90
df.loc[filt,'합산'] = df.loc[filt,'합산']+100
df
```

데이터 삭제
- columns
- index

drop - columns


```python
# 데이터 삭제 - drop :columns, 적용 : inplace =True
df.drop(columns=['합산'],inplace=True)

# 컬럼 2개 삭제
df.drop(columns=['국어','영어'])

# 전체 컬럼의 슬라이싱을 가지고 특정 컬럼을 삭제할 수 있ㅇ므
df.columns[3:5]
df.drop(columns=df.columns[3:5])
```

```python
df.drop(columns=['합계','평균','결과'])
```

drop - index 삭제

```python
# drop 삭제 - index(row 값 삭제) : inplace=True 반영
df.drop(index='1번')
df.head(1)

```

```python
# '김'이 포함된 학생을 삭제
filt = df['이름'].str.contains('김')
filt 

df.drop(index=df[filt].index)

# 수학 점수가 80점 미만 삭제
filt = df['수학']<80
filt

# 조건식을 이용한 삭제
df.drop(index=df[filt].index)
```

```python
filt = df['수학']<80
filt

# 조건식을 이용한 삭제
df.drop(index=df[filt].index)
```

```python
# row 추가 - index 이름 추가해서 입력하면 가능, 입력형태 list 입력하면  됨
df.loc['9번'] = ['유승찬','디지털고등학교',184,100,100,100,100,'Django',100*4,400/4]
df
```
