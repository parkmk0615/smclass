```python
import pandas as pd

df=pd.read_csv('score.csv',index_col='지원번호')
df
```

```python


# 학교 구로고 -> 가산고, 디지털고 -> 단지고로 변경
# sw특기 소문자로변경
# 4번sw특기 -> django로 변경
# 5번 sw 특기 -> python으로 변경
# 합계 , 결과 컬럼 추가
# 결과 컬럼은 합계 컬럼이 300점 이상이면 pass, 아니면 Fail

df['학교'].replace({'구로고':'가산고','디지털고':'단지고'},inplace=True)

df['SW특기']=df['SW특기'].str.lower()


df.loc['4번','SW특기'] = 'django'
df.loc['5번','SW특기'] = 'python'


df['합계']=df['국어']+df['영어']+df['수학']+df['사회']


df['결과']='Fail'
filt=df['합계']>=300
df.loc[filt,'결과']='pass'


df
```
