```python
### 함수
def func(value):
  if pd.notnull(value) :
    return value.lower()
  else:
    return 'python'
# 함수 호출 : apply()
df['SW특기']=df['SW특기'].apply(func)
df
```

```python
# 1. 요일 컬럼 추가
# 2. 0부터 6까지의  숫자를 랜덤으로 입력
# 3. 함수를 사용해서 0: 일요일, 1: 월요일.......6: 토요일로 변경해서 출력

import random

df['요일'] = 0

#filt= df['요일'] == 0
#df.loc['1번','요일'] = random.randint(0,6)
#df.iloc[0,8] =random.randint(0,6)
for i in range(8):
  df.iloc[i,8] = random.randint(0,6)

def func(value):
  if value == 0:
    return "일요일"
  elif value == 1:
    return "월요일"
  elif value == 2:
    return "화요일"
  elif value == 3:
    return "수요일"
  elif value == 4:
    return "목요일"
  elif value == 5:
    return "금요일"
  elif value == 6:
    return "토요일"
    

df['요일'] = df['요일'].apply(func)

df
```
