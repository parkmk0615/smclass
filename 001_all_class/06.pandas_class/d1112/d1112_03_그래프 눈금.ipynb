```python
import matplotlib.pyplot as plt
import matplotlib
# 한글
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size']= 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# x,y축 데이터
x=[1,2,3] # 리스트,튜플,DataFrame데이터  - 리스트 형태
y=[2,4,8]


# 선그래프 생성
plt.plot(x,y)

# 상단 타이틀
plt.title("라인그래프 Line",size=20)
# color : 폰트 색상, loc : 글자 위치

# 그래프 축 설명
plt.xlabel('X축 날짜',color='red',loc='center') # 그래프 X축 설명 / loc =right,left,center
plt.ylabel('Y축 인원',color='blue',loc='center') # 그래프 Y축 설명 / loc = top, center, bottom

# 그래프 눈금 단위 표시
plt.xticks([1,2,3])
plt.yticks([2,4,8,10])

plt.show()

```
