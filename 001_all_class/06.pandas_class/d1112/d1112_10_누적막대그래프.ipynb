```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size']= 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx')
```

- 누적막대그래프
- bottom : 첫번째 값을 입력

```python
plt.figure(figsize=(7,4))
plt.bar(df['이름'],df['국어'],label='국어')
plt.bar(df['이름'],df['영어'],bottom=df['국어'],label='영어')
plt.bar(df['이름'],df['수학'],bottom=df['국어']+df['영어'],label='수학')
plt.xticks(rotation=45)
plt.legend(ncol=3, loc='upper center')
plt.ylim(0,350)
for i in range(8):
  plt.text(df['이름'][i], df['국어'][i] + 0.2, df['국어'][i], color='blue',ha='center')
  plt.text(df['이름'][i], df['국어'][i] + df['영어'][i], df['영어'][i], color='blue',ha='center')
  plt.text(df['이름'][i], df['국어'][i] + df['영어'][i] + df['수학'][i], df['수학'][i],color='blue',ha='center')

plt.show()
```
