```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size']= 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('출산율2.xlsx', skiprows=2, nrows=2, index_col=0)
df
```

```python
# 출산율 그래프 출력
plt.figure(figsize=(10,5))
plt.title('출산율')
plt.plot(df.columns,df.iloc[0],marker ='o', markeredgecolor = 'purple', color='pink', label='출산율')
for idx in range(10):
  plt.text(df.columns[idx], df.iloc[0][idx]-15, df.iloc[0][idx], ha='center')

plt.legend()
plt.show()
```
