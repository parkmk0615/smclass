```python
# 원그래프
values = [30,25,20,13,10,2]

# edgecolor: 테두리 색, linewidth: 테두리 두께
wedgeprops={'width':0.6,'edgecolor':'black','linewidth':2}
labels=['python','java','javascript','oracle','html','etc']
colors=['plum','beige','skyblue','bisque','khaki','peachpuff']

## 데이터 값이 2%이하는 출력되지 않도록 함수적용
# pct : values의 % 값 생성
# def custom_autopct(pct):
#   return f'{pct:.2f}%' if pct>=10 else f'{pct:.0f}'
def custom_autopct(pct):
  if pct >=10 :
    #return ('%.1f%%' % pct)
    #return ('{:.1f}%'.format(pct))
    return (f'{pct:.1f}%')
  else:
    return ''


# wedgeprops: 중간 공백 / pctdistance : 글자 위치
plt.pie(values, labels=labels, colors=colors, startangle=90, autopct=custom_autopct, wedgeprops=wedgeprops, counterclock=False,pctdistance=0.7)
plt.show()
```
