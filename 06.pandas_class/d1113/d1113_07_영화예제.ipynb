```python
import pandas as pd
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
# 평균 추가
df.loc[8]=['평균',2024,df['관객 수'].mean(),df['평점'].mean()]
```

```python
### 영화 관객수를 막대 그래프로 출력하시오
### 평균 막대 그래프를 추가하고, 평균 선을 그어서 표시하시오 
### 이름 : 평균, 년도: 2024, 관객수 : 평균 관객수, 평점, 평균평점

# 평균 행 추가
df.loc[8]=['평균',2024,df['관객 수'].mean(),df['평점'].mean()]

movie=df['영화']
people=df['관객 수']
a = df['관객 수'].mean()
plt.figure(figsize=(10,5))
plt.grid(axis='y',color='purple',ls='--',linewidth=2,alpha=0.2)  # 격자선

plt.plot(df.index,[a]*9,ls='--', color='red') 

plt.ylim(1200,1800) # y축 범위 표시 1200~1800

s_bar=plt.bar(movie,people)
s_bar[len(df.index)-1].set_color('red') #평균 그래프 막대바 색깔 변경

plt.xticks(rotation=90)

plt.show()
```
