```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size']= 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx')
```

다중 막대 그래프 - 옆으로 여러개 막대그래프


```python
df.shape
```

```python
# 배열[0,1,2,3,4,5,6,7]
index=np.arange(df.shape[0])
index
```

```python
w=0.25
plt.bar(index,df['국어'],width=w,label='국어')
plt.bar(index-w,df['영어'],width=w,label='영어')
plt.bar(index+w,df['수학'],width=w,label='영어')
plt.xticks(index,rotation=45)
plt.ylim(0,120)

## for 값을 그래프에 출력
for i in range(8):
  plt.text(index[i], df['국어'][i], df['국어'][i], ha='center')
  plt.text(index[i]-w, df['영어'][i], df['영어'][i], ha='center')
  plt.text(index[i]+w, df['수학'][i], df['수학'][i], ha='center')

plt.legend(ncol=3,loc='upper center')
plt.show()
```
