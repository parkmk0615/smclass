```python
import pandas as pd
import numpy as np

df= pd.read_csv('chipotle.tsv',sep="\t")
df
```

```python
### 메뉴별 주문 수량을 출력하시오
df.groupby('item_name')['quantity'].count()
### 메뉴별 총주문량을 출력하시오
df.groupby('item_name')['quantity'].sum()
### 주문당 평균 계산 금액을 출력하시오
df['item_price'] = df['item_price'].str[1:].astype(float)
df.groupby('order_id')['item_price'].sum().mean()
### 아이템 별 가격 금액을 출력하시오
df.groupby('item_name')['item_price'].sum()
### 주문별, 높은 가격순으로 주문 개수를 출력하시오
df.groupby('order_id')['item_price'].sum().sort_values(ascending=False)
```
