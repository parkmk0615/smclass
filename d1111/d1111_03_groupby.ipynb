```python
import pandas as pd
import numpy as np
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
df.groupby('학교').count()
```

```python
# 그룹화
# 해당되는 그룹의 리스트를 보여줌
# df.groupby('학교').get_group('구로고')
# df.groupby('학교').mean()
df.groupby('학교').size() # 해당 학교 학생 수 확인
df.groupby('학교').size()['구로고'] # 해당 학교 학생 수 확인
df.groupby('학교').size()['가산고'] # 해당 학교 학생 수 확인  
df.groupby('학교')['키'].mean() # 학교별 키의 평균 값 출력
```

그룹별 최대값, 최소값,평균 출력

```python
# 학교 그룹별 키의 평균 출력
df.groupby('학교')['키'].mean()

# 학교 그룹별 키와 국어점수의 평균 출력
df.groupby('학교')[['키','국어']].mean()

# 학교 그룹별 키의 최대값 출력
df.groupby('학교')['키'].max()
```

```python
# 컬럼별, 최대값,최소값,평균,합계,표준편차
df['키'].mean()
```

```python
# 그룹화기준
# 컬럼을 기준으로 그룹핑 sum() : 합계 , count() : 개수, size() : 크기
df.groupby('학교').count()

# 컬럼 지정 필요 : df.groupoby('컬럼')[숫자타입컬럼] mean(),max(),min()
df.groupby('학교')['키'].min()
df.groupby('학교')[['키','국어']].min()

# 그룹별 키 순차 정렬
df.groupby('학년')['키'].mean().sort_values

# 그룹별 키로 정렬, ascending=False : 역순정렬
df.groupby('학년')['키'].mean().sort_values(ascending=False)

# valu_counts()
df.groupby('학교')['키'].value_counts()
```

```python
# 학교로 그룹핑 학년으로 학생 수 출력
## 그룹핑 했을때 사용되는 개수
df.groupby('학교')['학년'].value_counts()
df
```

```python
# 학교로 그룹핑을 해서 구로고 학새만 학년으로 학생 수 출력
df.groupby('학교')['학년'].value_counts().loc['구로고']

# 퍼센트로 출력됨 : nomalize=True
df.groupby('학교')['학년'].value_counts(normalize=True).loc['구로고']
```

```python
# 구로고 학생의 키의 합계를 구하시오
df.groupby('학교')['키'].sum()
```

```python
df.groupby('학교')['키'].mean()
df.groupby('학교')['키'].min()
df.groupby('학교')['키'].max()
df.groupby('학교')['키'].sum()
df.groupby('학교')['키'].agg(['mean','min','max','sum'])
```
