```python
import matplotlib.pyplot as plt
import matplotlib
# 한글
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size']= 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# x,y축 데이터
x=[1,2,3] # 리스트,튜플,DataFrame데이터  - 리스트 형태
y=[2,4,8]
yy=[1,8,3]


# 선그래프 생성
# linewidth : 선 두께, color : 선 색상
# marker :지점 표시 - o,x,v / markersize : 마커 크기, markeredgecolor : 마커 테두리 선 색깔, markerfacecolor : 마커색상
# alpha :투명도, ls : 라인 스타일 ':' :점선, '-' : dash선,  '-.':실선+점
# b: 선색상, v: 마커표시, ':': 선모양
plt.plot(x,yy,'bv:',label ='비용 변동',marker ='o', markersize=10, markeredgecolor = 'red')
plt.plot(x,y,label ='인원 변동',linewidth=5, marker ='v', markersize=10 , markerfacecolor='yellow',alpha=0.4, ls = '-.')

# 상단 타이틀
plt.title("라인그래프 Line",size=20)

# 범례 : 그래프 선 설명 문구 / loc : 범례 위치  upper, center, lower, right, left, 
# plt.legend(loc='upper center')
plt.legend(loc=(0.3,0.1)) # 좌표값으로 위치 설정
```
