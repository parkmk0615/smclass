```python
import matplotlib.pyplot as plt
import matplotlib
# 한글
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size']= 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

그래프에 값 표시

```python
x=[1,2,3] # 리스트,튜플,DataFrame데이터  - 리스트 형태
y=[2,4,8]

plt.figure(figsize=(7,5))
plt.plot(x,y,label='범례데이터',marker="o")
# text(x축 좌표위치,y축 좌표위치,데이터 값) , ha:글자 가운데 정렬 
#plt.text(x[0],y[0]+0.2,y[0],ha='center')
#plt.text(x[1],y[1]+0.2,y[1],ha='center')

## for  
for idx,values in enumerate(y):
  plt.text(x[idx]+0.05,y[idx]-0.25,y[idx],ha='center',color='blue')

plt.title('라인그래프')
plt.legend()
plt.show()
```
