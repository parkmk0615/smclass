```python
import matplotlib.pyplot as plt
import matplotlib
# 한글
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size']= 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
## 라벨 변경, 라벨의 기울기 설정, wudth, alpha 투명도 조정
```

```python
label = ['홍길동','유관순','강감찬']
values=[180,187,184]
colors = ['r','g','b']
plt.figure(figsize=(5,3))
plt.bar(label,values,color=colors,width=0.2,label='학생키') 
plt.yticks([0,50,100,150,200,250,300])
plt.xticks(label,['1번','2번','3번'], rotation=45) # x라벨 변경
# width : 바 크기 조절
plt.title('막대그래프') # 상단 제목
plt.legend(loc=('upper center'))




#plt.bar(label,values,color='orange')
#plt.plot(label,values) # 선 그래프
```

```python
colors=['red','green','blue']
bar = plt.bar(label,values,width=0.5,color=colors,alpha=0.5)
# 막대 그래프의 무늬설정 
bar[0].set_hatch('/')
bar[1].set_hatch('X') 
bar[2].set_hatch('.') 
```

- 0부터 범위 설정 xticks, yticks 
- 툭정 구간부터 범위 설정  xlim, ylim

```python

colors=['red','green','blue']
bar = plt.bar(label,values,width=0.5,color=colors,alpha=0.5)
#plt.yticks([0,100,150,200]) # 0부터 범위지정
plt.ylim(175,190) # 특정 범위지정
plt.show()
```

```python

label = ['홍길동','유관순','강감찬']
values=[180,187,184]
colors=['red','green','blue']
plt.barh(label,values,color=colors, height=0.5)
plt.xlim(175,190)

```
