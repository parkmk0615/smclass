```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size']= 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df.index
```

```python
plt.title('학생성적그래프')
plt.plot(df.index,df['국어'],label='국어')
plt.plot(df.index,df['영어'],label='영어')
plt.plot(df.index,df['수학'],label='수학')
plt.legend()

for i in range(8):
  plt.text(df.index[i],df['국어'][i]+0.6,df['국어'][i],ha='center')
  plt.text(df.index[i],df['영어'][i]+0.6,df['영어'][i],ha='center')
  plt.text(df.index[i],df['수학'][i]+0.6,df['수학'][i],ha='center')

plt.show()
```

```python
def n_func(x):
  if x-50>0:
    return x
  else:
    return 0

## apply : pandas 사용문법
n2_list = df['국어'].apply(n_func)
print(n2_list)
```

```python
def func(x):
  if x-30<0:
    return 0
  return x-30
```

```python
#  누적막대 그래프
plt.figure(figsize=(8,4))
plt.title('학생성적그래프')
plt.bar(df['이름'],df['국어'],width=0.6,label='국어')
plt.bar(df['이름'],df['영어'], bottom=df['국어'],width=0.6,label='영어')
plt.bar(df['이름'],df['수학'], bottom=df['국어']+df['영어'],width=0.6,label='수학')
n_list= df['국어'].apply(func)
plt.legend()

for i in range(8):
  plt.text(df['이름'][i], n_list[i], df['국어'][i], color='blue', ha='center')
  plt.text(df['이름'][i], df['국어'][i] + df['영어'][i]-30, df['영어'][i], color='blue', ha='center')
  plt.text(df['이름'][i], df['국어'][i] + df['영어'][i] + df['수학'][i]-30, df['수학'][i], color='blue', ha='center')

```
