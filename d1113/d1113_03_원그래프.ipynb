```python
# 원그래프
values = [1,1,1,1,2,1]
explode=[0.1,0.1,0,0,0,0]
labels=['python','java','javascript','oracle','html','etc']
colors=['plum','beige','skyblue','bisque','khaki','peachpuff']
# pie함수(데이터 값), labels : 글자출력
# explode : 그래프 분리, counterclock=False: 시계방향으로 
# startangle = 90 : 12시부터 시작됨.
# # 데이터 값 출력 : autopct
plt.title('프로그램 사용률')
plt.pie(values, labels=labels, explode=explode, autopct='% .1f%%', counterclock=False, startangle=180, colors=colors)
plt.legend(loc=(1.2,0.3),title='언어별 선호도')
plt.show()
```
