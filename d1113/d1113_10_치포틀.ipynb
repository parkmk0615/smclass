```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size']= 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False

df = pd.read_excel('score.xlsx',index_col='지원번호')
#df=pd.read_csv('drinks.csv')
```

```python
df = pd.read_csv('chipotle.tsv',sep="\t")

df
```

```python
# 메뉴별 주문 개수를 10개만 뽑아서 막대그래프로 출력

order = df.groupby('item_name')['order_id'].count().head(10)
order.index

plt.figure(figsize=(10,5))
plt.bar(order.index,order)
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=45)

# 평균을 추가하여 그래프로 표시
order['mean'] = df.groupby('item_name')['order_id'].count().mean()

plt.grid(axis='y',alpha=0.5)

order_bar=plt.bar(order.index,order)
order_bar[len(order.index)-1].set_color('red')

plt.plot(order.index,[order.loc['mean']]*len(order.index),color='black',ls='--')

plt.show()
```
