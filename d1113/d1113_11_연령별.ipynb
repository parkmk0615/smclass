```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size']= 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 남자인구 현황
df = pd.read_excel('연령별인구현황_202410.xlsx', skiprows=3, usecols='B,E:Y', nrows=1, index_col='행정기관')
df 

# 여자인구 현황
df2 = pd.read_excel('연령별인구현황_202410.xlsx', skiprows=3, usecols='B,AB:AV', nrows=1, index_col='행정기관')
df2
```

```python
# 남자 
# 천단위 변경, int 타입으로 변경
df.iloc[0] = df.iloc[0].str.replace(",","").astype(int)

# 여자 
# 천단위 변경, int 타입으로 변경
df2.iloc[0] = df2.iloc[0].str.replace(",","").astype(int)
```

```python
df.loc['전국  '][0]
```

```python
### 남자 컬럼, 여자 컬럼 복사
df2.columns = df.columns
df2
```

```python

plt.figure(figsize=(25,10))
plt.title('2024년 인구현황')
plt.barh(df.columns,-df.iloc[0]//1000,label='남자')
plt.barh(df2.columns,df2.iloc[0]//1000,label='여자')
plt.xlabel('인구수')
plt.ylabel('연령대')

plt.xticks(rotation=45)


plt.legend()
plt.show()
```
